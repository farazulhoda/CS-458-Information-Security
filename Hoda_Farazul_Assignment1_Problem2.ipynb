```python
import networkx as nx
import matplotlib.pyplot as plt

from collections import defaultdict
```

```python
def degree_anonymization(d, k):
    n = len(d)
    d_hat = [0] * n
    count = defaultdict(int)
    
    for i in range(n):
        count[d[i]] += 1
    
    for i in range(n):
        if d[i] >= k:
            d_hat[i] = d[i]
        else:
            d_hat[i] = k
            count[d[i]] -= 1
            if count[d[i]] == 0:
                del count[d[i]]
            while k in count:
                k += 1
    
    return d_hat
```

```python
def generate_random_graph(n, m):
    G = nx.gnm_random_graph(n, m)
    nx.draw(G, with_labels=True)
    plt.show()

n_values = [6, 10, 20]
for n in n_values:
    m = int(1.5 * n)
    generate_random_graph(n, m)
```

```python
def sort_degree_sequence(G):
    degrees = dict(G.degree())
    degree_sequence = sorted(degrees.values(), reverse=True)
    return degree_sequence

for n in [6, 10, 20]:
    m = int(1.5 * n)
    G = nx.gnm_random_graph(n, m)
    degree_sequence = sort_degree_sequence(G)
    print(degree_sequence)
```
